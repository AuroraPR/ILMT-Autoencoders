```python
from datetime import datetime
from types import SimpleNamespace
import paho.mqtt.client as mqtt
import ast # para convertir string a tipo list
import time
import json
import numpy as np
import os
import cv2
from IPython.display import clear_output

pathBinary=os.path.join(os.getcwd(),"data/binary")
pathEnergy=os.path.join(os.getcwd(),"data/energy")
pathThermal=os.path.join(os.getcwd(),"data/thermal")
pathWearable=os.path.join(os.getcwd(),"data/wearable")

broker = "192.168.1.38"

def getTime():
    return str(int(time.time()))+"\t"+datetime.now().strftime("%Y-%m-%d %H:%M:%S")

def key_file_time():
    return str((int)(time.time()/(60*60*24)))

def key_file_time_now():
    return str((int)(time.time()))
    
def on_message(client, userdata, msg):

    
    print (msg.topic + " " + str(msg.payload,'utf-8'))
    clear_output(wait=True)
    if msg.topic == "/amaltea/image/termal/2":
        
        print("termal data")
        image = np.frombuffer(msg.payload, dtype=np.float64)
        print(image.shape)
        image = np.reshape(image, (120, 160))
        print(image.shape)
        cv2.imwrite(os.path.join(pathThermal , key_file_time_now()+"T.jpg"), image.astype(np.uint8))
        file = open(os.path.join(pathThermal , key_file_time()+"."+"thermal"+".tsv"), 'a')
        file.write(getTime()+"\t"+key_file_time_now()+"T.jpg"+"\n")
        file.close()
        
    if msg.topic.startswith("/amaltea/wearable/ble/rssi/"):
        device=msg.topic.replace("/amaltea/wearable/ble/rssi/","")
        #print("device:",device)
        file = open(os.path.join(pathWearable , key_file_time()+"."+device+".tsv"), 'a')
        data = str(msg.payload,'utf-8')
        #print("rssi data:",data)
        rssis = json.loads(data)
        for rssidata in rssis:
            print(rssidata)
            rssie=json.loads(str(rssidata).replace("\'", "\""))
            #print(rssie['rssi'])
            file.write(str(int(time.time()))+"\t"+str(rssie['rssi'])+"\t"+str(rssie['addr'])+"\n")
        file.close()
        
    if msg.topic.startswith("shellies/amaltea/emeter/0"):
        device=msg.topic.replace("shellies/amaltea/emeter/0/","")
        #print("device:",device)
        data = str(msg.payload,'utf-8')        
        #print("data:",data)
        file = open(os.path.join(pathEnergy , key_file_time()+"."+device+".tsv"), 'a')
        file.write(getTime()+"\t"+str(data)+"\n")
        file.close()
    if msg.topic.startswith("amaltea/binary_sensor/") and msg.topic.endswith("/state"):
        device=msg.topic.replace("amaltea/binary_sensor/","").replace("/state","")
        #print("device:",device)
        data = str(msg.payload,'utf-8')        
        #print("data:",data)      
        file = open(os.path.join(pathBinary , key_file_time()+"."+device+".tsv"), 'a')
        file.write(getTime()+"\t"+str(data)+"\n")
        file.close()
    if msg.topic.startswith("amaltea/sensor/light") and msg.topic.endswith("/state"):
        device=msg.topic.replace("amaltea/sensor/","").replace("/state","")
        #print("device:",device)
        data = str(msg.payload,'utf-8')        
        #print("data:",data)      
        file = open(os.path.join(pathBinary , key_file_time()+"."+device+".tsv"), 'a')
        file.write(getTime()+"\t"+str(data)+"\n")
        file.close()    
    if msg.topic.startswith("amaltea/sensor/hum_") and msg.topic.endswith("/state"):
        device=msg.topic.replace("amaltea/sensor/","").replace("/state","")
        #print("device:",device)
        data = str(msg.payload,'utf-8')        
        #print("data:",data)      
        file = open(os.path.join(pathBinary , key_file_time()+"."+device+".tsv"), 'a')
        file.write(getTime()+"\t"+str(data)+"\n")
        file.close()
    if msg.topic.startswith("amaltea/sensor/") and msg.topic.endswith("_current_consumption/state"):
        device=msg.topic.replace("amaltea/sensor/","").replace("/state","")
        #print("device:",device)
        data = str(msg.payload,'utf-8')        
        #print("data:",data)      
        file = open(os.path.join(pathBinary , key_file_time()+"."+device+".tsv"), 'a')
        file.write(getTime()+"\t"+str(data)+"\n")
        file.close()
    if msg.topic.startswith("amaltea/sensor/temp_") and msg.topic.endswith("/state"):
        device=msg.topic.replace("amaltea/sensor/","").replace("/state","")
        #print("device:",device)
        data = str(msg.payload,'utf-8')        
        #print("data:",data)      
        file = open(os.path.join(pathBinary , key_file_time()+"."+device+".tsv"), 'a')
        file.write(getTime()+"\t"+str(data)+"\n")
        file.close()          
        
def on_connect(self, mosq, obj, rc):
        print('Connected with result code ' + str(rc))
        # Subscribing in on_connect() means that if we lose the connection and
        # reconnect then subscriptions will be renewed. There are other methods to achieve this..
        #client.subscribe("homeassistant/binary_sensor/#", 0)
        #print("Suscrito a topic homeassistant/binary_sensor/# ")     
        
def on_publish(client, obj, mid):
    None
    
def on_subscribe(mqttc, obj, mid, granted_qos):
    print("Subscribed: " + str(mid) + " " + str(granted_qos))

if __name__ == '__main__':

    client = mqtt.Client()
    client.on_message = on_message
    client.on_connect = on_connect
    client.on_publish = on_publish
    client.on_subscribe = on_subscribe

    client.connect(broker,1883,5)

    client.subscribe("#", 0)


    while True:
        client.loop()    
```
